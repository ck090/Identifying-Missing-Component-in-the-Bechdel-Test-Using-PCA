```python
#HERE dict3 should contain the list of all the line_ids and their cosine distance to tthe top 3 genres
tokenizer = RegexpTokenizer(r'\w+')

dict3 = {}
gender_dict = {}
index1 = 0

# For a single dialogue
for i in range(len(result_df)):
    line=ml_df.loc[ml_df['Line_Id']==result_df['line_id'][i]]['Dialogue'].iloc[0]
#     print (line)
    try:
        line=tokenizer.tokenize(line)
    except:
        print (line)
        line=''
    index1=index1+1
    av=np.zeros(300)
    val2={}
    try:
        line=[x.lower() for x in line]
    except:
        line=['']
    line=[w for w in line if not w in stop_words]
    l1=len(line)
    for tokens in line:
        try:
            av+=model[tokens]
        except:
            l1-=1
            pass
    av=av/l1

    #Finding Cosine distance for each of top 3 genres of that movie and storing it corresponding to that line_id in dict3
    for g in final_genres[result_df['movie'][i]]:

        try:
            val2[g]= 1 - spatial.distance.cosine(av, dict1[g])
        except:
            pass
    dict3[result_df['line_id'][i]]=val2

print (dict3)        
```
